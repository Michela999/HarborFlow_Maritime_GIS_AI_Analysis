```python
import requests
import pandas as pd

lat = 43.55
lon = 10.30

url = (
    "https://archive-api.open-meteo.com/v1/archive"
    f"?latitude={lat}&longitude={lon}"
    "&start_date=2023-10-01"
    "&end_date=2023-10-31"
    "&hourly=temperature_2m,wind_speed_10m,wind_direction_10m,"
    "surface_pressure,precipitation,weather_code"
    "&timezone=UTC"
)

response = requests.get(url)
data = response.json()

df = pd.DataFrame(data["hourly"])
df["latitude"] = lat
df["longitude"] = lon

# save using absolute path
output_path = r"C:\Users\miche\Documents\HarborFlow_Fresh\data\raw\weather\weather_openmeteo_livorno_oct2023.csv"
df.to_csv(output_path, index=False)

df.head()
```
